```python
# Function to filter and return columns containing a specific substring
def columns(df, j):
    """
    Filters the DataFrame and returns columns whose names contain the substring 'j'.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    j (str): The substring to search for in column names.

    Returns:
    pd.DataFrame: A DataFrame containing only the filtered columns.
    """
    return df.loc[:, df.columns.str.contains(j, case=False, na=False)]
```

```python

import pandas as pd  # Importing pandas for data manipulation

# Initialize an empty list to store the results of the grouped medians
p_list = []

# Loop through columns that contain 'return' in their name
for i in columns(data, 'return').columns:

    # Grouping the data by 'Date_y' (year) and 'YearQuarterPeriod', then calculating the median for the current column 'i'
    temp = data.groupby(['Date_y', 'YearQuarterPeriod'])[i].median()

    # Store the result of each grouped median in the list
    p_list.append(temp)

# Concatenate all results from the list into a single DataFrame
p = pd.concat(p_list, axis=1)

# Displaying the final DataFrame that contains median return values for each year-quarter and each 'return' column
print(p)
```

```python
# Function to filter columns based on a specific year and string pattern in the column name
def corr_year(df, j, k, k1):
    """
    Filters columns in the DataFrame based on a specific year and a string pattern in the column name.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    j (str): The column name representing the year (e.g., 'Date_y').
    k (int): The specific year to filter.
    k1 (str): The substring to match in the column names.

    Returns:
    pd.DataFrame: A DataFrame containing only the columns matching the criteria.
    """
    d = []  # List to store the filtered columns

    # Loop through all columns in the DataFrame
    for i in df:
        if k1 in i:  # Check if the column name contains the substring k1
            # Filter rows where the year (in column j) matches the specified year (k)
            d.append(df[df[j] == k][i])

    # Concatenate the filtered columns along axis 1 (horizontally) to form a new DataFrame
    d = pd.concat(d, axis=1)

    return d  # Return the filtered DataFrame
```

```python
epsilon = 1e-5  # To prevent division by zero

def combo_metric(m, n, suffix):
    a = df.columns[df.columns.str.contains(m)]
    b = df.columns[df.columns.str.contains(n)]

    for i, j in zip(a, b):
        if i.endswith(suffix) and j.endswith(suffix):
            ratio_ij = df[i] / (df[j] + epsilon)
            ratio_ji = df[j] / (df[i] + epsilon)

            col_name_ij = f'{i} / {j}'
            col_name_ji = f'{j} / {i}'

            df[col_name_ij] = ratio_ij
            df[col_name_ji] = ratio_ji

            print(f'Added: {col_name_ij}')
            print(f'Added: {col_name_ji}')
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def check_vif(data, threshold=10.0):
    """
    Checks Variance Inflation Factor (VIF) for multicollinearity.

    Args:
        data (pd.DataFrame): Input features (X) for regression.
        threshold (float): VIF threshold (default=5.0).
                          - VIF > 5 → Moderate multicollinearity
                          - VIF > 10 → Severe multicollinearity

    Returns:
        pd.DataFrame: VIF results with conditional formatting.
    """
    vif_data = pd.DataFrame()
    vif_data["Feature"] = data.columns
    vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]

    # Conditional formatting for multicollinearity warning
    vif_data["Multicollinearity"] = vif_data["VIF"].apply(
        lambda x: "✅ Low" if x < threshold else "⚠️ Moderate" if x < 10 else "❌ High"
    )

    return vif_data.sort_values("VIF", ascending=False)

# Example Usage:
# df_features = pd.DataFrame(...)  # Your feature matrix (X)
# vif_results = check_vif(df_features)
# print(vif_results)
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def check_vif_below_threshold(data, threshold=10.0):
    """
    Returns only features with VIF below the given threshold.

    Args:
        data (pd.DataFrame): Input features (X) for regression.
        threshold (float): VIF cutoff (default=5.0).

    Returns:
        pd.DataFrame: Features with VIF < threshold, sorted by VIF.
    """
    vif_data = pd.DataFrame()
    vif_data["Feature"] = data.columns
    vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]

    # Filter features below threshold and sort
    low_vif_features = vif_data[vif_data["VIF"] < threshold].sort_values("VIF")

    return low_vif_features

# Example Usage:
# df_features = pd.DataFrame(...)  # Your feature matrix (X)
# good_features = check_vif_below_threshold(df_features, threshold=5.0)
# print(good_features)
```
